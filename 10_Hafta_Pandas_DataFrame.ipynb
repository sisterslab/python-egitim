**Pandas DataFrame**

Satırları ve sütunları olan iki boyutlu etiketli veri yapısıdır. Excelin tamamı olarak düşünebiliriz.

**DataFrame Oluşturma**

pandas.DataFrame(data, index, columns)

* data -> Veri kümemiz
* index -> zorunlu alan değildir. varsayılan olarak 0 dan başlar n adet veri için n-1 indexine kadar devam eder.
* columns -> Opsiyoneldir. Veri için sütun isimlerini tanımlamamız için kullanılır.

```python
import pandas as pd
```

Boş bir DataFrame oluşturma

```python
df = pd.DataFrame()

print(df)
```

```python
df = pd.DataFrame(["Paris", "Berlin", "Ankara", "Atina"])

df
```

```python
df = pd.DataFrame(["Paris", "Berlin", "Ankara", "Atina"], columns=["Başkentler"])

df
```

Bir dictionary'den DataFrame oluşturma

```python
data = { "Food": ["Ekmek", "Dondurma", "Elma", "Bisküvi"],
        "Calories":[100, 430, 90, 120] }

df = pd.DataFrame(data)

df
```

index değerleri ekleyelim

```python
data = { "Food": ["Ekmek", "Dondurma", "Elma", "Bisküvi"],
        "Calories":[100, 430, 90, 120] }

df = pd.DataFrame(data, index=["food1", "food2", "food3", "food4"])

df
```

Sütun isimlerini değiştirme

```python
df = pd.read_csv("movies.csv")

df.columns = ["puan", "film_adi", "izleyici_kitlesi", "tur", "sure", "oyuncu_listesi"]

df.columns
```

sütun isimlerini birleştirme

```python
df["tur_film"] = df["tur"] + ", " + df.film_adi

df
```

Sıralama - sort_values() metodu kullanılır.

tersten sıralama için ascending=False kullanılır.

```python
df.film_adi.sort_values()
```

```python
df.film_adi.sort_values(ascending = False)
```

```python
df.sort_values("film_adi")
```

```python
df
```

```python
film_sure = []

for i in df.sure:
    if i >= 120: 
        film_sure.append("Çok Uzun")
    elif i < 120 and i >= 90:
        film_sure.append("Normal")
    else:
        film_sure.append("Kısa")

df["film_sure"] = film_sure

df.head()
```

filtreleme

```python
df[df.film_sure == "Normal"]
```

```python
df[(df.film_sure == "Normal") & (df.tur == "Drama") | (df.puan > 0)]
```

string işlemleri

```python
df["film_adi_kucuk"] = df.film_adi.str.lower()

df
```
